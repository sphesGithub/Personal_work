```python
#Data Aquisition
# Function to load datasets



def load_dataset(dataset_name):
    if dataset_name.endswith('.csv'):
        return pd.read_csv(dataset_name), dataset_name
    else:
        datasets = {
            'breast_cancer': load_breast_cancer,
            'diabetes': load_diabetes,
            'digits': load_digits,
            'iris': load_iris,
            'wine': load_wine
        }
        if dataset_name in datasets:
            data_bunch = datasets[dataset_name]()
            if hasattr(data_bunch, 'target'):
                data = pd.DataFrame(data_bunch.data, columns=data_bunch.feature_names)
                data['target'] = data_bunch.target
            else:
                raise ValueError(f"Dataset '{dataset_name}' does not have a target attribute.")
            return data, dataset_name
        else:
            raise ValueError("Invalid dataset name or format. Please choose one of: breast_cancer, diabetes, digits, iris, wine, or provide a valid CSV file.")

# (breastCancerData,breastCancer_names)=load_dataset("breast_cancer")
# breastCancerData
```

```python
#Data sampling

def sampler(data):
    """ 
    samples[0]=X_train
    samples[1]=X_test
    samples[2]=y_train
    samples[3]=y_test
    samples[4]=X
    samples[5]=Y

    """
    samples=[]
    X = data.drop(columns=['target'])  # Extract features
    y = data['target']  # Extract target
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    samples.append(X_train)
    samples.append(X_test)
    samples.append(y_train)
    samples.append(y_test)
    samples.append(X)
    samples.append(y)
    return samples
# #Now we have the splited data
# for i in range(6):
#     val=sampler(breastCancerData)[i]
#     print(val.shape)
```

```python
#EDA
def eda(X, y, plot_types):
    # Combine X and y into a single DataFrame for easier plotting
    data = X.copy()
    data['target'] = y

    for plot_type in plot_types:
        if plot_type == 'heatmap':
            plt.figure(figsize=(10, 8))
            sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
            plt.title('Correlation Heatmap')
            plt.show()

        elif plot_type == 'pairplot':
            sns.pairplot(data, hue='target', diag_kind='kde')
            plt.title('Pairplot')
            plt.show()

        elif plot_type == 'histogram':
            data.hist(bins=30, figsize=(20, 15))
            plt.suptitle('Histograms')
            plt.show()

        elif plot_type == 'boxplot':
            plt.figure(figsize=(12, 8))
            sns.boxplot(data=data)
            plt.title('Boxplot')
            plt.xticks(rotation=90)
            plt.show()

        elif plot_type == 'violinplot':
            plt.figure(figsize=(12, 8))
            sns.violinplot(data=data, x='target', y=data.columns[0], inner='quartile')
            plt.title(f'Violin Plot of {data.columns[0]}')
            plt.show()

        elif plot_type == 'scatter_matrix':
            pd.plotting.scatter_matrix(data, alpha=0.2, figsize=(20, 15), diagonal='kde')
            plt.suptitle('Scatter Matrix')
            plt.show()

        elif plot_type == 'countplot':
            plt.figure(figsize=(10, 6))
            sns.countplot(x='target', data=data)
            plt.title('Count Plot of Target Variable')
            plt.show()

        elif plot_type == 'distplot':
            for column in data.columns:
                if column != 'target':
                    plt.figure(figsize=(10, 6))
                    sns.histplot(data[column], kde=True)
                    plt.title(f'Distribution Plot of {column}')
                    plt.show()

        elif plot_type == 'jointplot':
            if data.shape[1] > 2:
                plt.figure(figsize=(10, 6))
                sns.jointplot(data=data, x=data.columns[0], y=data.columns[1], kind='scatter')
                plt.title(f'Joint Plot of {data.columns[0]} and {data.columns[1]}')
                plt.show()

        else:
            print(f"Plot type '{plot_type}' is not recognized. Please use a valid plot type.")
# plot_types = ['heatmap', 'pairplot', 'histogram', 'boxplot', 'violinplot', 'scatter_matrix', 'countplot', 'distplot', 'jointplot']
# ploter=["boxplot",'heatmap']
# eda(sampler(breastCancerData)[4], sampler(breastCancerData)[5], ploter)
```
